```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
columns = ['batch_score']
df = pd.read_csv("./logunet_simo_cls.csv", usecols=columns)
df.plot()
plt.xlabel("Epoch")
plt.ylabel("Score")
plt.show()
```

```python
df1 = pd.read_csv("./logunet_simo_cls.csv")
df2 = pd.read_csv("./logunet_miso_t.csv")
df3 = pd.read_csv("./logunet_miso_f.csv")
df4 = pd.read_csv("./logunet_miso_t128.csv")
# Plot a simple line chart
plt.plot(df1['epoch'], df1['batch_score'], color='g', label='SIMO')
# Plot another line on the same chart/graph
plt.plot(df2['epoch'], df2['batch_score'], color='r', label='MISO T')
plt.plot(df3['epoch'], df3['batch_score'], color='b', label='MISO F recadrée')
plt.plot(df4['epoch'], df4['batch_score'], color='c', label='MISO F centralisée')
plt.xlabel("Epoch")
plt.ylabel("Score")
plt.legend()
plt.show()
```
